```python
import matplotlib.pyplot as plt
import numpy as np
plt.title("line plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
xpoints=np.array([0,2])
ypoints=np.array([1,2])
plt.plot(xpoints,ypoints)
plt.show()
```

```python
plt.plot([5,200],[5,5])
plt.show()
#x1 X2 Y1 Y2
```

```python
plt.plot([5,200],[5,5],"--")
plt.show()
```

```python
x=[1,2,3]
y=[1,2,3]
plt.scatter(x,y)
plt.show()
```

```python
xpoints=np.array([1,2,6,8])
plt.plot(xpoints,marker="*")
plt.show()
```

```python
xpoints=np.array([1,2,6,8])
plt.plot(xpoints,marker="*",color='r')
plt.show()
```

```python
xpoints=np.array([1,2,6,8])
plt.plot(xpoints,marker="*",color='r',linewidth='15',mec='g')
plt.show()
```
