```python
import matplotlib.pyplot as plt
country=["INDIA","USA","UK","FRANCE"]
gdp=["10000","30000","20000","20000000"]
plt.bar(country,gdp,color="palegreen")
plt.title("county vs gdp")
plt.xlabel('country')
plt.ylabel('gdp')
plt.show()
```

```python
import matplotlib.pyplot as plt
country=["INDIA","USA","UK","FRANCE"]
gdp=["1000","3000","20000","2000"]
plt.barh(country,gdp,color="palegreen")
plt.title("county vs gdp")
plt.ylabel('gdp')
plt.xlabel('country')
plt.show()
```
