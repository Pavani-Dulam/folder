```python
#1Read Total profit of all months and show it using a line plot
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("company.csv")
profitlist=df["total_profit"].tolist()
monthlist=df["month_number"].tolist()
plt.plot(monthlist,profitlist,label="month wise profit data")
plt.xlabel('month number')
plt.ylabel('profit in dollar')
plt.xticks(monthlist)
plt.title("company profit per month")
plt.yticks([100000,200000,300000,400000,500000])
plt.legend(loc="upper left")
plt.show()
```

```python
#2Get total profit of all months and show line plot with the following Style properties
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("company.csv")
profitList = df ['total_profit'].tolist()
monthList  = df ['month_number'].tolist()

plt.plot(monthList, profitList, label = 'Profit data of last year', 
      color='r', marker='o', markerfacecolor='k', 
      linestyle='--', linewidth=3)
      
plt.xlabel('Month Number')
plt.ylabel('Profit in dollar')
plt.legend(loc='lower right')
plt.title('Company Sales data of last year')
plt.xticks(monthList)
plt.yticks([100000, 200000, 300000, 400000, 500000])
plt.show()
```

```python
 #3Read all product sales data and show it  using a multiline plot
import pandas as pd
import matplotlib.pyplot as plt  
df=pd.read_csv("company.csv")
monthList=df['month_number'].tolist()
face=df['facecream'].tolist()
wash=df['facewash'].tolist()
tooth=df['toothpaste'].tolist()
bath=df['bathingsoap'].tolist()
shampoo=df['shampoo'].tolist()
moisturizer=df ['moisturizer'].tolist()
plt.plot(monthList,face,label='Face cream Sales Data',marker='o',linewidth=3)
plt.plot(monthList,wash,label='Face Wash Sales Data',marker='o',linewidth=3)
plt.plot(monthList,tooth,label='ToothPaste Sales Data',marker='o',linewidth=3)
plt.plot(monthList,bath,label='ToothPaste Sales Data',marker='o',linewidth=3)
plt.plot(monthList,shampoo,label='ToothPaste Sales Data',marker='o',linewidth=3)
plt.plot(monthList,moisturizer,label='ToothPaste Sales Data',marker='o',linewidth=3)
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.legend(loc='upper left')
plt.xticks(monthList)
plt.yticks([1000, 2000, 4000, 6000, 8000, 10000, 12000, 15000, 18000])
plt.title('Sales data')
plt.show()
```

```python
#4read toothpaste sales data of each month and show it using a scatter plot
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("company.csv")
monthList=df['month_number'].tolist()
pastesales=df['toothpaste'].tolist()
plt.scatter(monthList,pastesales,label='tooth paste sales data')
plt.xlabel('month number')
plt.ylabel('no of units')
plt.legend(loc='upper right')
plt.title("tooth paste sales per month")
plt.grid(True,linewidth=1,linestyle="--")
plt.show()
```

```python
#5read face cream and facewash,product sales data and show it using bar chart
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("company.csv")
monthList=df['month_number'].tolist()
creamsales=df['facecream'].tolist()
washdata=df['facewash'].tolist()
plt.bar([a-0.5 for a in monthList],creamsales,width=0.25,label='face cream sales data',align='edge')
plt.bar([a+2.5 for a in monthList],washdata,width=-0.25,label='face cream sales data',align='edge')
plt.xlabel('month number')
plt.ylabel('sales units')
plt.legend(loc='upper left')
plt.title("sales data of facewash and facecream")
plt.xticks(monthList)
plt.grid(True,linewidth=1,linestyle="--")
plt.show()



```

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("company.csv")
monthList=df['month_number'].tolist()
bathsoap=df['bathingsoap'].tolist()
facewashdata=df['facewash'].tolist()
axarr=plt.subplots(2,sharex=True)
axarr[0].plot(monthList,bathsoap,label='bathing soap sales data',color='k',marker='o',linewidth=3)
axarr[0].set_title("bathsoap sales data")
axarr[1].plot(monthList,facewashdata,label='facewash sales data',color='b',marker='o',linewidth=3)
axarr[1].set_title("facewash sales data")
plt.xticks(monthList)
plt.xlabel('month number')
plt.ylabel('sales in units')
plt.show()
```

```python
#9read bathing soap facewash of all months and diplay suing sub plot
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("company.csv")
monthList=df['month_number'].tolist()
bathsoap=df['bathingsoap'].tolist()
facewashdata=df['facewash'].tolist()
plt.subplot(2,1,1)
plt.title('bathsoap sales')
plt.plot(monthList,bathsoap,label='bathsoap sales data',color='g',marker='o')
plt.legend(loc='upper left')
plt.xlabel('month number')
plt.ylabel('sales units')
plt.subplot(2,1,2)
plt.title('facewash sales')
plt.plot(monthList,facewashdata,label='facewash sales data',color='r',marker='o')
plt.legend(loc="upper right")
plt.xticks(monthList)
plt.xlabel('month number')
plt.ylabel('sales units')
plt.show()
```

```python
#10 read all products sales data and show it using the stack plot
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("company.csv")
monthList=df['month_number'].tolist()
creamdata=df['facecream'].tolist()
wash=df["facewash"].tolist()
tooth=df['toothpaste'].tolist()
bath=df["bathingsoap"].tolist()
sham=df['shampoo'].tolist()
mois=df['moisturizer'].tolist()
plt.plot([],[],color='m',label='face cream',linewidth=5)
plt.plot([],[],color='c',label='face wash',linewidth=5)
plt.plot([],[],color='r',label='tooth paste',linewidth=5)
plt.plot([],[],color='k',label='bathing soap',linewidth=5)
plt.plot([],[],color='g',label='shampoo',linewidth=5)
plt.plot([],[],color='y',label='moisturizer',linewidth=5)
plt.stackplot(monthList,creamdata,wash,tooth,bath,sham,mois,colors=['m','c','r','k','g','y'])
plt.xlabel("month number")
plt.ylabel("sales units")
plt.title("all product sales data using stack plot")
plt.legend(loc="upper right")
plt.show()
```
